## Step 1 - Import Packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

## Step 2 - Data Loading and Inspection
Load the Iris dataset into a pandas DataFrame

```python
iris = load_iris()
df = pd.DataFrame(data=iris.data,columns=iris.feature_names)
df
```

## Step 4 - Data Visualization

### Histogram for one numerical feature (petal length)

```python
# Create a figure with a specific size
plt.figure(figsize=(10, 6))

# Create histogram for petal length
# Note: The exact column name in the iris dataset is 'petal length (cm)'
sns.histplot(data=df, x='petal length (cm)', kde=True, color='purple', bins=15)

# Add title and labels
plt.title('Distribution of Petal Length', fontsize=15)
plt.xlabel('Petal Length (cm)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

### Box plot to showing the distribution and potential outliers for one feature, grouped by the flower class (species).

```python
# Add species column to the DataFrame
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Create a figure with a specific size
plt.figure(figsize=(12, 7))

# Create box plot for petal length grouped by species (updated syntax)
sns.boxplot(x='species', y='petal length (cm)', hue='species', data=df, palette='viridis', legend=False)

# Add swarm plot to show individual data points
sns.swarmplot(x='species', y='petal length (cm)', data=df, color='black', alpha=0.5, size=4)

# Add title and labels
plt.title('Distribution of Petal Length by Species', fontsize=16)
plt.xlabel('Species', fontsize=14)
plt.ylabel('Petal Length (cm)', fontsize=14)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7, axis='y')

# Enhance the appearance
plt.tight_layout()
plt.show()
```

###  Scatter plot of petal length vs. petal width, with points colored by the flower species, to visually assess separability

```python
# Create scatter plot of petal length vs. petal width, colored by species
sns.scatterplot(
    x='petal length (cm)', 
    y='petal width (cm)', 
    hue='species',
    data=df,
    palette='viridis',
    s=100,  # Point size
    alpha=0.8,  # Transparency
    edgecolor='white',  # White edge around points
    linewidth=0.5
)

# Add title and labels
plt.title('Petal Length vs. Petal Width by Species', fontsize=16)
plt.xlabel('Petal Length (cm)', fontsize=14)
plt.ylabel('Petal Width (cm)', fontsize=14)

# Customize the legend
plt.legend(title='Species', fontsize=12, title_fontsize=14)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Add a tight layout
plt.tight_layout()

# Show the plot
plt.show()
```
